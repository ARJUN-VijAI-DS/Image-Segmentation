<font size = 5.5> **What is Image Segmentation?**

<font size = 4.5> Image segmentation using K-Means clustering based on pixel colour values aims to partition the image into regions with similar colours.
    <br><br>This process simplifies the image and can be useful for various purposes.

<font size = 5>**Practical Applications**
    <font size = 4.5><br><br> **Object Recognition:** Identifying and isolating objects or regions within an image, such as distinguishing different parts of a scene or identifying specific objects.
    <br><br> **Preprocessing for Further Analysis:** Serving as a **preprocessing step** for more complex analysis, such as object detection or classification.
    <br><br> **Image Compression:** Reducing the number of colors can help in compressing the image data while maintaining the essential features.
    <br><br> **Medical Imaging:** Highlighting specific regions of interest, such as tumors or organs, based on color differences in medical scans.

<font size = 4.5>**An image** is a visual representation of information, captured or created in a way that can be displayed and perceived visually. <br><br>It is composed of tiny elements called pixels, which are the smallest units of an image. Each pixel contains colour and intensity information, which together create the overall visual content.

<font size = 5.5>**Key Characteristics of an Image**
    <br><br>**Pixels:**
    <br>- **The basic unit of an image.**
    <br> - Each pixel has a specific colour and intensity.
    <br> - In digital images, pixels are arranged in a grid format.
    <br><br>**Resolution:**
    <br>- Refers to the number of pixels in an image, typically described as width x height (e.g., 1920x1080).
    <br>- **Higher resolution means more pixels, leading to greater detail and clarity.**
    <br><br>**Colour Depth:**
    <br>- The number of bits used to represent the color of each pixel.
    <br>- Common color depths include 8-bit (256 colors), 16-bit, **24-bit (True Color, which can display about 16.7 million colors)**, and higher.
    <br><br>**Colour Models:**
    <br>- **RGB: Red, Green, Blue** channels combine to create a wide array of colors. Common for digital screens.
    <br>- **CMYK: Cyan, Magenta, Yellow, Key (Black)**. Used in color printing.
    <br>- **Grayscale: Single-channel image representing shades of gray** - Black and White.
    <br>- **HSV: Hue, Saturation, Value**. Useful for certain image processing tasks.

<font size = 4.5> **How are pixels represented numerically?**
    <br><br>Pixels can be numerically represented in various ways depending on the type of image and the colour model used. 


|Image Type|	Pixel Representation|	Example|
|--------|----|----|
|**Grayscale - (Black and white)**|	Single value (0-255)|	128 (Medium gray)
|**RGB**|	Three values (0-255)|	[255, 0, 0] (Red)
|**HSV**|	Three values (H:0-360, S:0-100, V:0-100)|	[0, 100, 100] (Bright red)
|**CMYK**|	Four values (0-100%)|	[0, 100, 100, 0] (Pure red)
|**Lab**|	Three values (L*:0-100, a*: -128 to 127, b*: -128 to 127)|	[50, 70, 20] (Specific colour)
|**Binary**|	Single value (0 or 1)|	1 (White)

```python
import numpy as np
from PIL import Image
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
# Convert the image to a numpy array
image_np = np.array(image)
image_np
```

<font size = 4.5>**Shape of the image (height, width, 3)**
    <br><br>- Height(720): The number of rows of pixels in the image. It measures how tall the image is.
    <br><br>- Width(1280): The number of columns of pixels in the image. It measures how wide the image is.
    <br><br>- 3: Represents the three color channels (Red, Green, Blue).

```python
# Reshape the image to a 2D array of pixels
pixel_values = image_np.reshape((-1, 3))
print(pixel_values)
```

```python
pixel_values.shape
```

```python
# Define the number of clusters (K)
k = 20
# Apply KMeans
kmeans = KMeans(n_clusters=k, random_state=42)
labels = kmeans.fit_predict(pixel_values)
```

```python
len(labels)
```

```python
centers = np.uint8(kmeans.cluster_centers_)

# Map the labels to the center points (colors)
segmented_image = centers[labels.flatten()]
```

```python
# Reshape back to the original image dimensions
segmented_image = segmented_image.reshape(image_np.shape)
```

```python
# Display the original and segmented images using matplotlib
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Original image
ax1.imshow(image)
ax1.set_title('Original Image')
ax1.axis('off')

# Segmented image
ax2.imshow(segmented_image)
ax2.set_title('Segmented Image with KMeans')
ax2.axis('off')

plt.show()
```
